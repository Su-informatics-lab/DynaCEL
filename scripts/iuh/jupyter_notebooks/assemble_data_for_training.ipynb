```python
import os 
import pandas as pd
```

```python
raw_data_dir = "data/raw/iuh/IUH DATA/CSV Files/"
processed_data_dir = "data/processed/iuh"
```

```python
demographics = pd.read_csv(os.path.join(processed_data_dir, "demographics.csv"))
demographics["ArriveDTS"] = pd.to_datetime(demographics["ArriveDTS"])
demographics["DischargeDTS"] = pd.to_datetime(demographics["DischargeDTS"])
demographics
```

```python
demographics.columns
```

```python
demographics = demographics[["MRN", "FIN", "Deceased", "DOB", "Race", "Ethnicity", "Sex", "ht", "wt", "bmi_cal", "HTN_patient", "ArriveDTS", "DischargeDTS"]]
```

```python
demographics["death_status"] = 1
demographics.loc[demographics["Deceased"].isna(), "death_status"] = 0
```

```python
vitals = pd.read_csv(os.path.join(raw_data_dir, "Vitals.csv"))
vitals["EventDT"] = pd.to_datetime(vitals["EventDT"])
vitals
```

```python
hr = vitals[vitals["Event"].str.contains("heart rate", case=False)]
hr
```

```python
sbp = vitals[vitals["Event"].str.contains("systolic blood pressure", case=False)]
sbp
```

```python
merged_hr = pd.merge(hr, demographics, on=["MRN", "FIN"], how="left")
merged_hr["time_diff"] = merged_hr["EventDT"] - merged_hr["ArriveDTS"]
merged_hr
```

```python
merged_hr.to_csv(os.path.join("data/data_for_training/iuh", "merged_hr_events.csv"), index=False)
```

```python
merged_sbp = pd.merge(sbp, demographics, on=["MRN", "FIN"], how="left")
merged_sbp["time_diff"] = merged_sbp["EventDT"] - merged_sbp["ArriveDTS"]
merged_sbp
```

```python
merged_sbp.to_csv(os.path.join("data/data_for_training/iuh", "merged_sbp_events.csv"), index=False)
```
